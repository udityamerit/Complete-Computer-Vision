```python
import os
import cv2 as cv
```

```python
# Load the video
video = cv.VideoCapture('../assets/Social-distance.mp4')
```

```python
# Display the video
while True:
        ret, frame = video.read()   ## ret return the True and False value on the video loading  and frame get the array of video

        if not ret:
                break
        
        cv.imshow('video frames', frame)
        if cv.waitKey(10) & 0xFF == ord('q'):
                break

video.release()
cv.destroyAllWindows()
```

```python
# Write a video

video = cv.VideoCapture('../assets/Social-distance.mp4')

fourcc = cv.VideoWriter_fourcc(*'mp4v')

width = int(video.get(cv.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv.CAP_PROP_FRAME_HEIGHT))
out = cv.VideoWriter('../outputs/new_video.mp4', fourcc, 30, (width, height))

while True:
        ret, frame = video.read()
        if not ret:
                break

        out.write(frame)

out.release()
video.release()
cv.destroyAllWindows()
```

```python
# Write a video

video = cv.VideoCapture('../assets/Social-distance.mp4')

fourcc = cv.VideoWriter_fourcc(*'mp4v')

width = int(video.get(cv.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv.CAP_PROP_FRAME_HEIGHT))
out = cv.VideoWriter('../outputs/new_video1.mp4', fourcc, 30, (width, height))

while True:
        ret, frame = video.read()
        frame = (frame) + 20
        if not ret:
                break

        out.write(frame)

out.release()
video.release()
cv.destroyAllWindows()
```
