```python
import os
import cv2 as cv
```

```python
img = cv.imread('../assets/butterfly.jpg')
```

```python
if img is None:
        print("Error: could not loaded image")

else:
        print("image loaded..")
```

```python
# display the image

cv.imshow("image window", img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# write the image 
new_img = img+100
os.makedirs("../outputs", exist_ok=True)
```

```python
cv.imwrite('../outputs/new_img_butterfly.jpg', new_img)
```
